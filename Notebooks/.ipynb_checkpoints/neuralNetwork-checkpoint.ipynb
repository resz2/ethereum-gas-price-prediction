```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../Data/processed.csv')
df.head()
```

```python
dfu = pd.read_csv('../Data/unscaled.csv')
dfu.head()
```

```python
df.drop(['Unnamed: 0'], axis=1, inplace=True)
dfu.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
df.head()
```

## Train-test split

```python
from sklearn.model_selection import train_test_split
x = dfu
y = dfu['block_min']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

x_train_scaled = x_train.copy()
x_test_scaled = x_test.copy()
x_train_scaled[x_train_scaled.columns] = scaler.fit_transform(x_train_scaled[x_train_scaled.columns])
x_test_scaled[x_test_scaled.columns] = scaler.fit_transform(x_test_scaled[x_test_scaled.columns])
x_train_scaled.drop(['block_min'], axis=1, inplace=True)
x_test_scaled.drop(['block_min'], axis=1, inplace=True)
```

## Defining the model

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
dim = x_train_scaled.shape[1]
model = Sequential()
model.add(Dense(units=8, input_dim = dim, activation='relu', kernel_initializer='normal'))
model.add(Dense(units=32, activation='relu', kernel_initializer='normal'))
model.add(Dense(units=16, activation='relu', kernel_initializer='normal'))
model.add(Dense(units=1, kernel_initializer='normal'))
model.summary()
model.compile(optimizer = 'adam', loss = 'mean_squared_error', metrics = ['accuracy'])
```

## Training the model

```python
def train(model, x_train, y_train, bsize=10000, nepoch=64):
    for i in range(4):
        model.fit(x_train, y_train, batch_size = bsize, epochs = nepoch,verbose = 1)
        bsize //= 2
        nepoch //= 2
```

```python
train(model, x_train_scaled, y_train)
```

```python
model.evaluate(x_test_scaled, y_test)
```

```python
y_pred = model.predict(x_test_scaled)
y_p_tr = model.predict(x_train_scaled)
```

```python
def accuracy(y_pred,data):
    correct=0
    i=0
    for index,row in data.iterrows():
        if(y_pred[i]>=row['past_min'] and y_pred[i]<=row['past_max']):
            correct+=1
        i+=1
    acc = correct/data.shape[0]
    return acc 
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,accuracy_score, r2_score
from math import sqrt

def calculateMetric(y_pred, y_test, data):
    print("MAE=",mean_absolute_error(y_test,y_pred))
    print("RMSE=",sqrt(mean_squared_error(y_test,y_pred)))
    print("R2 Score", r2_score(y_test,y_pred))
    print('accuracy=', accuracy(y_pred, data))
```

```python
calculateMetric(y_pred, y_test, x_test)
```

```python
calculateMetric(y_p_tr, y_train, x_train)
```
